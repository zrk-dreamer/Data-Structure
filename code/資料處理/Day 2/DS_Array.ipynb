```python
#費氏數列
F = [0]*16
F[0] = 1
F[1] = 1
for i in range(2, 16):
    F[i] = F[i-1] + F[i-2]
for i in range(0, 16):
    print("F[",i,"] =",F[i])
```

```python
#『費氏級數』的副程式碼
def Fibonacci (n):
    if n<=1:
        return n 
    else:
        fnm2 = 0; fnm1 = 1
        for i in range(2,n+1):
            fn = fnm1 + fnm2
            fnm2 = fnm1
            fnm1 = fn
        return fn

for i in range(0, 16):
    print("F[",i,"] =",Fibonacci (i))
```

```python
#『費氏級數』的遞迴副程式碼
def FibonacciR (n):
    if n<=1:
        return n 
    else:
        return FibonacciR(n-1)+FibonacciR(n-2)

for i in range(0, 16):
    print("F[",i,"] =",FibonacciR (i))
```

```python
#多項式相加
#Ex. a(x)=3x^3+2x^2
terms = 2
TermA = [[0]*2]*terms
print(TermA)
TermA[0]=[3,3] #3x^3
TermA[1]=[2,2] #2x^2
print(TermA)

#Ex. b(x)=5x^4+x^2-2x+7
terms = 4
TermB = [[0]*2]*terms
print(TermB)
TermB[0]=[4,5] #5x^4
TermB[1]=[2,1] #x^2
TermB[2]=[1,-2] #-2x
TermB[3]=[0,7] #7
print(TermB)

def PolyAdd(A,B):
    aterms=len(A)
    bterms=len(B)
    TermR=[[0]*2]*1
    rterms=0
    Apos=Bpos=0
    while Apos<aterms and Bpos<bterms:
        if A[Apos][0]==B[Bpos][0]:            
            t=A[Apos][1]+B[Bpos][1]
            if t != 0:
                TermR[rterms][0]=A[Apos][0]
                TermR[rterms][1]=t
                TermR.append([0,0])
                rterms+=1
            Apos+=1
            Bpos+=1
        elif A[Apos][0]<B[Bpos][0]:
            TermR[rterms][0]=B[Bpos][0]
            TermR[rterms][1]=B[Bpos][1]
            TermR.append([0,0])
            rterms+=1
            Bpos+=1
        else:
            TermR[rterms][0]=A[Apos][0]
            TermR[rterms][1]=A[Apos][1]
            TermR.append([0,0])
            rterms+=1
            Apos+=1
    for i in range(Apos,aterms):
        TermR[rterms][0]=A[Apos][0]
        TermR[rterms][1]=A[Apos][1]
        TermR.append([0,0])
        rterms+=1
        Apos+=1
    for i in range(Bpos,bterms):
        TermR[rterms][0]=B[Bpos][0]
        TermR[rterms][1]=B[Bpos][1]
        TermR.append([0,0])
        rterms+=1
        Bpos+=1
    TermR.pop(-1)
    return TermR

TermC=PolyAdd(TermA,TermB)
Poly = 'c(x)= '
for i in range(len(TermC)): 
    Poly+=str(TermC[i][1])+"x^"+str(TermC[i][0])+"  +  "
print(Poly)
```

```python
#稀疏矩陣表示法與顯示
def PrintSparseMatrix(A):
    terms=len(A)-1
    rows=A[terms][0]
    cols=A[terms][1]
    MatrixA = str(rows)+"列,"+str(cols)+"行,"+str(terms)+"項非零值\n"
    CurrentA=0    
    for r in range(rows):
        for c in range(cols):
            if A[CurrentA][0]==r and A[CurrentA][1]==c:
                MatrixA+=str(A[CurrentA][2])+'\t'
                CurrentA+=1
            else:
                MatrixA+='0\t'
        MatrixA+='\n'
    print(MatrixA)
    
A=[[0]*3]*(8+1)
A[0]=[0,0,15]
A[1]=[0,3,22]
A[2]=[0,5,-15]
A[3]=[1,1,11]
A[4]=[1,2,3]
A[5]=[2,3,-6]
A[6]=[4,0,91]
A[7]=[5,2,28]
A[8]=[6,6,8] #6列6行8個非零元素

PrintSparseMatrix(A)
```

```python
#轉置矩陣
def SparseMatrixTranspose(A):
    terms=len(A)-1   
    T = [[0]*3 for i in range(terms+1)]           # A=[[0]*3]*(terms+1) 各項資料會連動,勿使用
    T[terms][1]=cols=A[terms][0]
    T[terms][0]=rows=A[terms][1]
    T[terms][2]=A[terms][2] 
    if terms > 0:
        CurrentT=0
        for c in range(cols): #根據行來轉置 
            for i in range(terms):
                if A[i][1] == c:                    
                    T[CurrentT][0]=c
                    T[CurrentT][1]=A[i][0]
                    T[CurrentT][2]=A[i][2]                    
                    CurrentT+=1        
    return T
```

```python
X=[[0]*3]*(8+1)
X[0]=[0,0,15]
X[1]=[0,3,22]
X[2]=[0,5,-15]
X[3]=[1,1,11]
X[4]=[1,2,3]
X[5]=[2,3,-6]
X[6]=[4,0,91]
X[7]=[5,2,28]
X[8]=[6,6,8] #6列6行8個非零元素

PrintSparseMatrix(X)
print()
PrintSparseMatrix(SparseMatrixTranspose(X))
```

```python
#快速轉置矩陣
def SparseMatrixFastTranspose(A):
    terms=len(A)-1
    T = [[0]*3 for i in range(terms+1)]
    T[terms][1]=cols=A[terms][0]
    T[terms][0]=rows=A[terms][1]
    T[terms][2]=A[terms][2] 
    if terms > 0:
        rowSize=[0]*cols
        rowStart=[0]*cols
        for i in range(terms):
            rowSize[A[i][1]]+=1
        rowStart[0]=0
        for i in range(1,cols):
            rowStart[i]=rowStart[i-1]+rowSize[i-1]
        for i in range(terms):
            j=rowStart[A[i][1]]
            T[j][0]=A[i][1]
            T[j][1]=A[i][0]
            T[j][2]=A[i][2]
            rowStart[A[i][1]]+=1
    return T
```
