```python
# 環狀鏈結串列 - 計算長度
class Node:
    def __init__(self, x):
        self.data = x
        self.next = None
class CirLinkedList: # 初始化
    def __init__(self):
        self.head = None
        self.last = None #加入最後一個元素指標
    def printCirLinkedList(self): # 列印 CirLinkedList
        CirLinkedList="head => "
        if self.head == None:
            CirLinkedList+='head (None)'            
        else:
            tmp = self.head
            while tmp != self.last:
                CirLinkedList+=str(tmp.data)+" => "
                tmp=tmp.next
            CirLinkedList+=str(tmp.data)+" => head"
        print(CirLinkedList)    
    def insertHead(self, x): # 前端插入元素
        newnodex=Node(x)
        if self.head == None:            
            newnodex.next = newnodex
            self.last = self.head = newnodex
        else:            
            newnodex.next = self.head
            self.last.next = newnodex            
            self.head = newnodex
    def insertBack(self, x): # 尾端插入元素
        newnodex=Node(x)
        if self.head == None:
            newnodex.next = newnodex
            self.last = self.head = newnodex
        else:
            newnodex.next=self.head
            self.last.next=newnodex
            self.last=newnodex
    def insert(self, y, x): # 在y後面插入x
        tmp = self.head
        while tmp!=self.last:
            if tmp.data == y:
                break
            tmp=tmp.next
        if tmp.data == y: #確保資料y存在
            newnodex=Node(x)
            if tmp==self.last:
                self.last=newnodex
            newnodex.next=tmp.next
            tmp.next=newnodex
    def removeHead(self): # 前端刪除元素
        if self.head != None: 
            if self.head != self.last:
                self.head = self.head.next
                self.last.next=self.head
            else:
                self.head = self.last = None
    def removeBack(self): # 尾端刪除元素
        if self.head != None:
            if self.head != self.last:
                tmp = self.head
                while tmp.next != self.last:
                    tmp = tmp.next
                self.last = tmp
                tmp.next = self.head
            else:
                self.head = self.last = None   
    def remove(self, x):
        if self.head != None:
            if self.head != self.last:                
                if self.head.data == x: #第一個元素是 x
                    self.head = self.head.next
                    self.last.next=self.head
                else:
                    tmp = self.head.next
                    before = self.head
                    while tmp != self.last:
                        if tmp.data == x:
                            break
                        before = tmp
                        tmp = tmp.next
                    if tmp.data == x:
                        if tmp == self.last: #最後一個元素是 x
                            self.last = before
                            before.next=self.head
                        else:
                            before.next=tmp.next
            elif self.head.data == x: #只剩一個元素而且是 x
                self.head = self.last = None
    def len(self): # 計算長度
        count = 0
        if self.head == None:
            return 0
        else:
            tmp = self.head
            while tmp.next != self.head:
                count += 1
                tmp = tmp.next                
            return count + 1
```

```python
#鏈結串列-連結兩個鏈結串列
class Node:
    def __init__(self, x):
        self.data = x    #Data
        self.next = None #Link
class LinkedList:
    def __init__(self):
        self.head = None
    def insertHead(self, x): # 前端插入元素
        if self.head == None:
            self.head = Node(x)
        else:
            newnodex=Node(x)
            newnodex.next = self.head
            self.head = newnodex
    def printLinkList(self):
        LinkListStr="head => "
        tmp = self.head
        while tmp != None:
            LinkListStr+=str(tmp.data)+" => "
            tmp=tmp.next
        LinkListStr+='None'
        print(LinkListStr)
    def insert(self, y, x): # 在y後面插入x
        tmp = self.head
        nodex = Node(x)
        while tmp != None:
            if tmp.data == y:
                break
            tmp = tmp.next
        nodex.next = tmp.next
        tmp.next = nodex
    def insertBack(self, x): # 尾端插入元素
        if self.head == None:
            self.head = Node(x)
        else:
            tmp = self.head
            while tmp.next != None:
                tmp = tmp.next
            tmp.next = Node(x)
    def removeHead(self): # 前端刪除元素
        if self.head != None:
            self.head = self.head.next
    def removeBack(self): # 尾端刪除元素
        if self.head != None:
            tmp = self.head
            while tmp.next.next != None:
                tmp = tmp.next
            tmp.next = None
    def remove(self, x):
        tmp =self.head
        while tmp != None:
            if tmp.data == x:
                break
            before = tmp
            tmp = tmp.next
        if tmp != None: #確保刪除資料存在
            if tmp == self.head:  #刪除第一個元素
                self.head = self.head.next
            else:
                before.next = tmp.next
    def Concatenate(self, LinkListB):
        if self.head == None:
            self.head=LinkListB.head
        else:
            tmp = self.head
            while tmp.next != None:
                tmp=tmp.next
            tmp.next=LinkListB.head
```

```python
#鏈結串列-反轉 ( Reverse )
class Node:
    def __init__(self, x):
        self.data = x    #Data
        self.next = None #Link
class LinkedList:
    def __init__(self):
        self.head = None
    def insertHead(self, x): # 前端插入元素
        if self.head == None:
            self.head = Node(x)
        else:
            newnodex=Node(x)
            newnodex.next = self.head
            self.head = newnodex
    def printLinkList(self):
        LinkListStr="head => "
        tmp = self.head
        while tmp != None:
            LinkListStr+=str(tmp.data)+" => "
            tmp=tmp.next
        LinkListStr+='None'
        print(LinkListStr)
    def insert(self, y, x): # 在y後面插入x
        tmp = self.head
        nodex = Node(x)
        while tmp != None:
            if tmp.data == y:
                break
            tmp = tmp.next
        nodex.next = tmp.next
        tmp.next = nodex
    def insertBack(self, x): # 尾端插入元素
        if self.head == None:
            self.head = Node(x)
        else:
            tmp = self.head
            while tmp.next != None:
                tmp = tmp.next
            tmp.next = Node(x)
    def removeHead(self): # 前端刪除元素
        if self.head != None:
            self.head = self.head.next
    def removeBack(self): # 尾端刪除元素
        if self.head != None:
            tmp = self.head
            while tmp.next.next != None:
                tmp = tmp.next
            tmp.next = None
    def remove(self, x):
        tmp =self.head
        while tmp != None:
            if tmp.data == x:
                break
            before = tmp
            tmp = tmp.next
        if tmp != None: #確保刪除資料存在
            if tmp == self.head:  #刪除第一個元素
                self.head = self.head.next
            else:
                before.next = tmp.next
    def Concatenate(self, LinkListB):
        if self.head == None:
            self.head=LinkListB.head
        else:
            tmp = self.head
            while tmp.next != None:
                tmp=tmp.next
            tmp.next=LinkListB.head
            
    def Reverse(self):
        current = self.head
        previous = None
        while current != None:
            r = previous
            previous = current
            current = current.next
            previous.next = r
        self.head = previous
```
